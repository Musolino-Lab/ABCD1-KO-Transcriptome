# Generate Axial Plots

```python
import numpy as np
import pandas as pd
from axial import axial
```

## I. Volcano & Bar

### Human

```python
human = pd.read_csv('../data/Yamuna/Human_NT_vs_ABCD1_DESeq2_results.tsv', sep='\t', header=None, names=['GeneID','Base mean','log2(FC)','StdErr','Wald-Stats','P-value','P-adj']).set_index('GeneID')
human['log2(FC)'] = human['log2(FC)'] * -1
human.head()
```

```python
human['q'] = -np.log10(human['P-adj'])
human_max_finite_log_qVal = np.max(human['q'][np.isfinite(human['q'])])
human['q'] = np.around(np.clip(human['q'], 0, human_max_finite_log_qVal), decimals=3)
print("max q-val: " + str(human_max_finite_log_qVal))

human['logFC'] = np.around(human['log2(FC)'], decimals=3)
human = human.dropna(how='any', subset=['logFC', 'q'])
```

```python
axial.volcano(human, 
              title='ABCD1 KO Transcriptome', 
              scripts_mode='CDN', 
              data_mode='inline', 
              organism='human', 
              output_dir='../docs', 
              filename='volcano.html', 
              version="latest")
```

```python
axial.bar(human, 
          title='ABCD1 KO Transcriptome', 
          scripts_mode='CDN', 
          data_mode='inline', 
          organism='human', 
          output_dir='../docs', 
          filename='bar.html', 
          version="latest")
```

## II. Heatmap


### Mouse, Human, Barres, and Lee Merged

```python
mouse = pd.read_csv('../data/Barres/Normalized_counts_Barres_mouse.tsv', sep='\t', index_col=0).round(3)
mouse = mouse[~(mouse.sum(axis=1) == 0)]
mouse = mouse.rename(columns={'Barres_Endothelial_Cells_rep1': 'Mouse_Barres_1', 'Barres_Endothelial_Cells_rep2': 'Mouse_Barres_2'})
mouse.head()
```

```python
mouse_classes = pd.DataFrame(mouse.columns.str.split('_').tolist(), columns=['system', 'condition', 'replicate'], index=mouse.columns)
del mouse_classes['replicate']
mouse_classes
```

```python
mouse.index = mouse.index.str.upper()
```

```python
merged = pd.concat([human, mouse, lee], axis=1, sort=True).dropna(how='all', subset=["Human_ABCD1_1", "Human_ABCD1_2", "Human_ABCD1_3", "Human_NT_1", "Human_NT_2", "Human_NT_3"])
merged.head()
```

```python
merged_classes = pd.concat([human_classes, mouse_classes, lee_classes]).replace('ccALD', 'ABCD1').replace('WT', 'NT')
merged_classes
```

```python
axial.heatmap(merged,
              merged_classes,
              title='ABCD1 KO Transcriptome', 
              scripts_mode='CDN', 
              data_mode='inline', 
              organism='human', 
              output_dir='../docs', 
              filename='heatmap.html', 
              version="latest")
```
