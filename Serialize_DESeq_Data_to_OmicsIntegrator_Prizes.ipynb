# Preparing prizes for OmicsIntegrator2 from DESeq2 results for Human

```python
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
NUM = 200
human = pd.read_csv('DESeq2_data/Yamuna/Human_NT_vs_ABCD1_DESeq2_results.tsv', sep='\t', header=None, names=['GeneID','Base mean','log2(FC)','StdErr','Wald-Stats','P-value','P-adj'])
human = pd.concat([human.set_index('GeneID').rename_axis(None).loc[['ABCD1', 'TGFB1']], human.set_index('GeneID').rename_axis(None)[:NUM]])
human.head()
```

```python
'CLDN5' in human.index.tolist(), 'ABCD1' in human.index.tolist(), 'TGFB1' in human.index.tolist(), 
```

```python
human.loc[human[human['P-value'] == 0].index.tolist(), 'P-value'] = 1e-309
human.loc[human[human['P-adj'] == 0].index.tolist(), 'P-adj'] = 1e-310
human.head()
```

```python
human['log2(FC)'] = human['log2(FC)'] * -1
```

```python
human.rename(columns={'Base mean': 'Base_Mean', 'log2(FC)': 'log2FC', 'StdErr': 'StdErr', 'Wald-Stats': 'Wald_Stats', 'P-value': 'P_value', 'P-adj': 'P_adj'}, inplace=True)
human.head()
```

```python
human['prize'] = -np.log10(human.P_adj) / 100
human.head()
```

```python
human.loc["CLDN5", "prize"] += 2
human.loc["ABCD1", "prize"] += 5
```

```python
human[['prize','Base_Mean','log2FC','StdErr','Wald_Stats','P_value','P_adj']].to_csv(f'/Users/alex/Documents/abcd1/human_deseq_prizes_{NUM}.tsv', sep='\t')
```
